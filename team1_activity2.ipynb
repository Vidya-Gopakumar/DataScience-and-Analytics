```python
import pandas as pd
import seaborn as sns
data=pd.read_csv('diabetes_model.csv')
diabetes_data=pd.DataFrame(data)
print(diabetes_data)
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
sns.countplot(diabetes_data['Outcome'])
```

```python
import matplotlib.pyplot as plt
sns.boxplot(x=diabetes_data['Outc'],y=diabetes_data['Age'],data=diabetes_data)
plt.title('effect of age on outcome')
plt.show()
```

```python
sns.violinplot(y=diabetes_data['DiabetesPedigreeFunction'],x=diabetes_data['Outcome'],data=diabetes_data)
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x=diabetes_data['Insulin'],y=diabetes_data['Glucose'],data=diabetes_data,hue=diabetes_data['Outcome'])
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(diabetes_data.corr(),annot=True)
plt.xticks(rotation=90)
```
