 1.Reading Data  in python environment

```python
import pandas as pd
data=pd.read_excel('iris.xls')
iris_data=pd.DataFrame(data)

```

2.Display the columns in the data set

```python
print(iris_data)
```

```python
3.Mean of each column of dataset
```

```python
mean_sl=iris_data['SL'].mean()
print("Mean of SL:",mean_sl)
mean_sw=iris_data['SW'].mean()
print("Mean of SW:",mean_sw)
mean_pl=iris_data['PL'].mean()
print("Mean of PL:",mean_pl)
mean_pw=iris_data['PW'].mean()
print("Mean of PW:",mean_pw)
```

4.Check for null values

```python
nul_value=pd.isnull(iris_data)
print(nul_value)
nul_value.sum()
```

5.Meaningfull visualization from iris data

  1.Pie chart of  iris classification distribution

```python
import matplotlib.pyplot as plt 
iris_pie =iris_data['Classification'].value_counts().rename_axis('Iris_Classification').reset_index(name='classification_count')
plt.figure(figsize=(5,5))
plt.pie(iris_pie.classification_count,labels=iris_pie.Iris_Classification,autopct = '%.2f%%')
plt.title('Iris Classification distribution')
plt.show()
```

 2.Violin plot to see the iris classification across PW values

```python
import seaborn as sns
sns.violinplot(y=iris_data['PW'],x=iris_data['Classification'],data=iris_data)
plt.ylabel('Petal width')
plt.show()
```

```python
 3.Comparison between PW and PL based on classification
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x=iris_data['PL'],y=iris_data['PW'],data=iris_data,hue=iris_data['Classification'],s=70)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

4.Comparison of SW and SL on the basis of classification

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x=iris_data['SL'],y=iris_data['SW'],data=iris_data,hue=iris_data['Classification'],s=70)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

5.swarm plot for the relation between petal length and classification

```python
import warnings
warnings.filterwarnings('ignore')
sns.swarmplot(x=iris_data['Classification'],y=iris_data['PL'],data=iris_data)
plt.ylabel('Petal Length')
plt.show()
```

6.Strip plot of classification based on sepal length(findings:one outlier in iris-virginica)

```python
sns.stripplot(x=iris_data['Classification'],y=iris_data['SL'],data=iris_data)
plt.xlabel('Classification')
plt.ylabel('Sepal Length')
plt.show()
```
